```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(columns=['Chapter No', 'Verse No', 'Shloka', 'English Translation', 'Explanation'])
```

```python
for chapter, n in zip(range(1, 19), [48, 73, 44, 43, 30, 48, 31, 29, 35, 43, 56, 21, 35, 28, 21, 25, 29, 79]):
    for verse in range(1, n):
        page1 = requests.get('https://bhagavadgita.io/chapter/{}/verse/{}'.format(chapter, verse))
        soup = BeautifulSoup(page1.content, "html.parser")
        h_text = soup.find("b").get_text().replace('\n', ' ').strip()
        e_text = soup.find("p", class_ = 'verse-meaning' ).get_text()
        # Explanation
        page2 = requests.get('https://www.holy-bhagavad-gita.org/chapter/{}/verse/{}'.format(chapter, verse))
        soup = BeautifulSoup(page2.content, "html.parser")
        try:
            explanation = soup.find(id = 'commentary').get_text().replace('\n', '').strip()
        except:
            explanation=np.nan
        df.loc[len(df.index)] = [chapter, verse, h_text, e_text, explanation]
```

```python
df.to_csv('Bhagvad Gita.csv', index=False)
```
